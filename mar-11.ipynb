```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
days = ["monday","tuesday","wednesday","thursday","friday","saturday"]
std = [3,4,5,7,11,12]
exp = [0,0.1,0.2,0.3,0,0.1]
# exp = np.ones(6)
# exp = np.zeros(6)
clr = ['r','b','k','c','m','g']
len(exp)
exp
```

```python
plt.pie(std, labels = days, explode = exp , colors = clr,
       autopct = "%0.1f%%", radius = 1.5)
plt.show()
```
