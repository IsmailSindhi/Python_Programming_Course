```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
classes = ["Python","R","AI","ML","DS"]
```

```python
class_one = [30,10,20,25,10]
class_two = [40,5,20,20,10]
class_three = [35,5,30,15,15]
```

```python
plt.bar(classes,class_one)
```

```python
plt.barh(classes,class_two)
```

```python
plt.bar(classes,class_one, width = 0.2, align = "edge",
        color = "y", edgecolor = "m", linewidth = 2, linestyle = "--")
plt.title("Bar chart example")
plt.xlabel("Classes", fontsize = 15)
plt.ylabel("No: of students", fontsize = 15)
```

```python
clasees_index = np.arange(len(classes))
width = 0.2
plt.barh(clasees_index, class_one, width, color ="b")
plt.barh(clasees_index + width, class_two, width, color = "g")
plt.barh(clasees_index + width + width, class_three, width, color ="y")
plt.yticks(clasees_index + width + width,classes , rotation = 20)
plt.xlabel("No of Students")
plt.ylabel("Classes")
plt.title("Bar Chart Example")
plt.show()
```
