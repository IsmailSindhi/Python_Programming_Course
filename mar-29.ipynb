```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df.info()
```

```python
l_enc = LabelEncoder()
df["smoker"] =  pd.DataFrame(l_enc.fit_transform(df['smoker']))
```

```python
df.describe()
```

```python
one_hot = OneHotEncoder(handle_unknown="ignore")
enc_df = pd.DataFrame(one_hot.fit_transform(df[["sex"]]).toarray())
```

```python
dum_df = pd.get_dummies(df[["sex"]])
```

```python
dum_df = pd.get_dummies(df[["region"]])
df = df.join(dum_df)
df
```

```python
df
```

```python
y = df["charges"]
```

```python
X = df.drop("charges",axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=74)
```

```python
model = LinearRegression()
model.fit(X_train,y_train)
```
